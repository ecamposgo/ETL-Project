```python
michelin_restaurants['MinPrice']=michelin_restaurants['MinPrice'].str.replace(',','')
michelin_restaurants['MaxPrice']=michelin_restaurants['MaxPrice'].str.replace(',','')

michelin_restaurants['MinPrice'] = pd.to_numeric(michelin_restaurants['MinPrice'],downcast='integer')
michelin_restaurants['MaxPrice'] = pd.to_numeric(michelin_restaurants['MaxPrice'],downcast='integer')
michelin_restaurants.info(memory_usage='deep')
```

```python
michelin_restaurants['Address'] = michelin_restaurants['Address'].str.rsplit(',',1).str[1]
```

```python
michelin_restaurants = michelin_restaurants.rename(columns={'Address': 'Country'})
michelin_restaurants.head()                                                         
```
