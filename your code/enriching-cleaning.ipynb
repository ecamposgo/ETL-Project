```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
michelin_restaurants = pd.read_csv('../data/michelin_restaurants.csv', encoding = 'latin1')
```

```python
michelin_restaurants.head()
```

```python
michelin_restaurants.shape
```

```python
michelin_restaurants.columns
```

```python
michelin_restaurants=michelin_restaurants[michelin_restaurants['award'] != 'Bib Gourmand']
michelin_restaurants.shape
```

```python
michelin_restaurants=michelin_restaurants.drop_duplicates(subset='name', keep='first')
michelin_restaurants.shape
```

```python
michelin_restaurants.info(memory_usage='deep')
```

```python
michelin_restaurants['minprice']=michelin_restaurants['minprice'].str.replace(',','')
michelin_restaurants['maxprice']=michelin_restaurants['maxprice'].str.replace(',','')

michelin_restaurants['minprice'] = pd.to_numeric(michelin_restaurants['minprice'],downcast='integer')
michelin_restaurants['maxprice'] = pd.to_numeric(michelin_restaurants['maxprice'],downcast='integer')
michelin_restaurants.info(memory_usage='deep')
```

```python
michelin_restaurants['address'] = michelin_restaurants['address'].str.rsplit(',',1).str[1]

michelin_restaurants['address']=michelin_restaurants['address'].str.strip()
```

```python
michelin_restaurants = michelin_restaurants.rename(columns={'name': 'restaurant','address': 'country'})
michelin_restaurants.head()                                                      
```

```python
eurorates = pd.read_csv('../data/eurorates.csv', encoding = 'latin1')
eurorates.head()
```

```python
michelin_restaurants=pd.merge(michelin_restaurants,eurorates,on="currency", how="left")
michelin_restaurants.head()
```

```python
michelin_restaurants['average_price'] = (michelin_restaurants['minprice'] + michelin_restaurants['maxprice'])/2
michelin_restaurants
```

```python
michelin_restaurants['avg_price_euro'] = (michelin_restaurants['average_price']/michelin_restaurants['exchange']).round(1)
```

```python
michelin_restaurants=michelin_restaurants.reindex(columns=['restaurant', 'award','country', 'location', 'minprice', 'maxprice', 
                                                           'currency','exchange','average_price','avg_price_euro', 'cuisine', 
                                                           'longitude', 'latitude', 'phonenumber','url', 'websiteurl', 
                                                           'facilitiesandservices'])
michelin_restaurants.head()
```
